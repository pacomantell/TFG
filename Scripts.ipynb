```python
def load(file):
    #CEAMA
    if(file==1):
        #PATH
        mes = ['1 ene', '2 feb', '3 mar', '4 abr', '5 may', '6 jun', '7 jul', '8 ago', '9 sep', '10 oct', '11 nov', '12 dic']
        path = ['CEAMA/2021 CEAMA Airly horarios '+mes[i]+'.xlsx' for i in range(len(mes))]
        
        #COLUMNAS
        colnames = ['To', 'PM1', 'PM2.5', 'PM10', 'Temperature', 'Humidity',  'Pressure', 'NO2', 'O3']
        col = [colnames[i]+'-CEAMA' for i in range(len(colnames))]

    #SANTO DOMINGO
    elif(file==2):
        #PATH
        path = ['SDomingo/2021 semana '+str(i)+' CRSDOM Airly horarios.xlsx' for i in range(1,53,1)]
    
        #COLUMNAS
        colnames = ['To', 'PM1', 'PM2.5', 'PM10', 'Temperature', 'Humidity',  'Pressure', 'SO2', 'CO']
        col = [colnames[i]+'-SDOMINGO' for i in range(len(colnames))]
    
    #LORCA
    elif(file==3):
        #PATH
        path = ['CentroLorca/2021 semana '+str(i)+' CLorca Airly horarios.xlsx' for i in range(1,53,1)]
        
        #COLUMNAS
        colnames= ['PM1', 'PM2.5', 'PM10', 'Temperature', 'Pressure', 'Humidity', 'NO₂', 'O₃']
        ID = ['ID 35828 - ', 'ID 43112 - ', 'ID 82480 - ']

    
    #lista con df por meses (cada mes un elemento de la lista)
    lista = [pd.read_excel(path[i], index_col=0) for i in range(len(path))]

    #combino todos los df en uno solo, de todo el año
    df = pd.concat(lista)
    
    #pongo los NaN como 0
    df.fillna(0,inplace=True)
    
    #renombro columnas
    if ((file==1) or (file==2)):
        df.columns = [col[i] for i in range(len(col))]
        
    #Para los de lorca unifico todos los ID
    elif(file==3):
        df_patch = pd.DataFrame(df['To'])
        for i in range(len(colnames)):
            colname = ['PM1', 'PM2.5', 'PM10', 'Temperature',  'Pressure', 'Humidity', 'NO2', 'O3']
            df_patch[colname[i]+'-LORCA'] = df[ID[0]+colnames[i]] + df[ID[1]+colnames[i]] + df[ID[2]+colnames[i]]
        df = df_patch

    #Sustituyo los 0 por NaN:
    df = df.replace(0.0, np.nan)
    return df
```

# SERIES TEMPORALES 

```python
#ESTA FUNCION ME CALCULA LAS MEDIAS POR DIA, MES, DIA DE LA SEMANA Y HORAS
def medias(df):
    
    #SEPARO JUNTA DE AIRLY PARA EL EXPORT
    if (df.columns[1].split('-')[1] in ['GRN', 'CONGR']):
        path = 'DATOS/JUNTA'
    else:
        path = 'DATOS/AIRLY'
    #Medias diarias (defino los dias en funcion de la estacion de medida)
    
    if (df.columns[1].split('-')[1] in ['CEAMA', 'GRN', 'CONGR']):
    
        dias = pd.date_range(start='2021-01-01 00:00:00', end='2022-01-01 00:00:00', freq='D').strftime('%Y-%m-%d')

        est_dia = [df.loc[(str(dias[i])+' 00:00:00'):(str(dias[i])+' 23:00:00')].describe() for i in range(len(dias)-1)]
    
    elif (df.columns[1].split('-')[1] in ['SDOMINGO', 'LORCA']):
        
        dias = pd.date_range(start='2021-01-04 00:00:00', end='2022-01-01 00:00:00', freq='D').strftime('%Y-%m-%d')
        
        est_dia = [df.loc[(str(dias[i])+' 00:00:00'):(str(dias[i])+' 23:00:00')].describe() for i in range(len(dias)-1)]
    
    df_dias = pd.concat(est_dia, keys=dias, names=['Fecha', 'datos'])
    
    
    #me quedo solo con las medias y las std
    medias_dias = df_dias.xs('mean', level=1)
    std_dias = df_dias.xs('std', level=1)
    
    #exporto a excel
    medias_dias.to_excel(path+'/medias_diarias_'+str(df.columns[1].split('-')[1])+'.xlsx')
    df_dias.to_excel(path+'/estadistica_diarias_'+str(df.columns[1].split('-')[1])+'.xlsx')
    
    #estadistica mensual
    mes = pd.date_range(start='2021-01', end='2022-01', freq='M').strftime('%Y-%m')
        
    est_mes = [df.loc[(str(mes[i]))].describe() for i in range(len(mes))]
    
    df_mes = pd.concat(est_mes, keys=mes, names=['Mes', 'datos'])
    
    #me quedo solo con las medias y las std
    media_mes = df_mes.xs('mean', level=1)
    std_mes = df_mes.xs('std', level=1)
    
    #exporto a excel
    media_mes.to_excel(path+'/medias_mes_'+str(df.columns[1].split('-')[1])+'.xlsx')
    df_mes.to_excel(path+'/medias_mes_'+str(df.columns[1].split('-')[1])+'.xlsx')
    
    #estadistica por dia de la semana (PARA TODO EL AÑO)
    
    #creo una columna con el dia de la semana
    dfw = df
    dfw['Dia'] = df.index.day_name()
    
    #hago la estadistica anual por dia de la semana
    df_week = dfw.groupby(['Dia']).describe().reindex(
        ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).T
    #me quedo con la media y la std
    media_semana = df_week.xs('mean', level=1).T
    std_semana = df_week.xs('std', level=1).T
    
    #exporto a excel
    df_week.to_excel(path+'/estadistica_dia_de_semana_'+str(df.columns[1].split('-')[1])+'.xlsx')
    media_semana.to_excel(path+'/media_dia_semana_'+str(df.columns[1].split('-')[1])+'.xlsx')
    
    #estadistica por hora del dia (PARA TODO EL AÑO)
    dfh = df
    dfh['Hora'] = df.index.hour
    
    est_horas = dfh.groupby(['Hora']).describe().T
    
    medias_horas = est_horas.xs('mean', level=1).T
    std_horas = est_horas.xs('std', level=1).T
    
    #exporto a excel
    est_horas.to_excel(path+'/estadistica_horas_año_'+str(df.columns[1].split('-')[1])+'.xlsx')
    medias_horas.to_excel(path+'/medias_por_hora_'+str(df.columns[1].split('-')[1])+'.xlsx')
    
    return medias_dias, std_dias, media_mes, std_mes, media_semana, std_semana, medias_horas, std_horas
```

```python
#FUNCION QUE ME OBTIENE LAS SERIES TEMPORALES DE UNA VARIABLE EN LAS DIFERENTES ESTACIONES PARA LOS VALORES ANUALES
def plot_año(df1, df2, df3, df4, var):
    fig, fig1 = plt.subplots(1,1, figsize=(12,6), dpi=200)
    dias = np.arange('2021-01-01', '2022-01-01', dtype='datetime64[D]')
    mes = np.arange('2021-01-15', '2022-01-15', dtype='datetime64[M]') + np.timedelta64(14, 'D')
    #DISTINGO RED AIRLY Y RED JUNTA
    
    #CODIGO DE COLORES EN FUNCION DE LA ESTACION
    #CEAMA/LORCA/SDOMINGO
    colors = ['violet', 'cornflowerblue', 'limegreen']
    colmes = ['purple', 'navy', 'forestgreen']
    #red airly
    if (df1.columns[0].split('-')[1] in ['CEAMA', 'SDOMINGO', 'LORCA']):
        
        path = 'SERIES/AIRLY/serie_'+str(var)+'_anual_AIRLY.eps'
        #Tengo en cuenta que no todas las variables estan en las tres estaciones
        if (var in ['CO', 'SO2']):
            #media diaria
            mean_day = df1[str(var)+'-SDOMINGO'].T.to_numpy()
            std_day = df2[str(var)+'-SDOMINGO'].T.to_numpy()
            labels = 'SDOMINGO'
        
            #media mensual
            mean_mes = df3[str(var)+'-SDOMINGO'].T.to_numpy()
            std_mes = df4[str(var)+'-SDOMINGO'].T.to_numpy()
        
            #plot
            fig1.plot(dias, mean_day, '.--', c=colors[2], lw=0.5, label=labels)
            fig1.errorbar(mes, mean_mes, fmt='h', c=colmes[2], yerr=(std_mes), capsize=4, label=str(labels)+'-mes')
    
        elif(var in ['O3', 'NO2']):
            #media diaria
            mean_day = df1[[str(var)+'-CEAMA', str(var)+'-LORCA']].T.to_numpy()
            std_day = df2[[str(var)+'-CEAMA', str(var)+'-LORCA']].T.to_numpy()
            labels = ['CEAMA', 'LORCA']
        
            #media mensual
            mean_mes = df3[[str(var)+'-CEAMA', str(var)+'-LORCA']].T.to_numpy()
            std_mes = df4[[str(var)+'-CEAMA', str(var)+'-LORCA']].T.to_numpy()
        
            #plot
            for i in range(len(mean_day)):
            
                fig1.plot(dias, mean_day[i], '.--', c=colors[i], lw=0.5, label=labels[i])
                fig1.errorbar(mes, mean_mes[i], fmt='h', c=colmes[i], yerr=(std_mes[i]),
                              capsize=4, label=str(labels[i])+'-mes')
    
        #PM1, PM2.5, PM10
        else:
            #media diaria
            mean_day = df1[[str(var)+'-CEAMA', str(var)+'-LORCA', str(var)+'-SDOMINGO']].T.to_numpy()
            std_day = df2[[str(var)+'-CEAMA', str(var)+'-LORCA', str(var)+'-SDOMINGO']].T.to_numpy()
            labels = ['CEAMA', 'LORCA', 'SDOMINGO']
        
        
            #media mensual
            mean_mes = df3[[str(var)+'-CEAMA', str(var)+'-LORCA', str(var)+'-SDOMINGO']].T.to_numpy()
            std_mes = df4[[str(var)+'-CEAMA', str(var)+'-LORCA', str(var)+'-SDOMINGO']].T.to_numpy()
        
            #plot
            for i in range(len(mean_day)):
                fig1.plot(dias, mean_day[i], '.--', c=colors[i], lw=0.5, label=labels[i])
                fig1.errorbar(mes, mean_mes[i], fmt='h', c=colmes[i], 
                              yerr=(std_mes[i]), capsize=4, label=str(labels[i])+'-mes')
            if (var != 'Pressure'):
                fig1.set_ylim(0,None)
    
    #red junta
    else:
        path = 'SERIES/JUNTA/serie_'+str(var)+'_anual_JUNTA.eps'
        
        if (var == 'O3'):
            #media diaria
            mean_day = df1[str(var)+'-CONGR'].T.to_numpy()
            std_day = df2[str(var)+'-CONGR'].T.to_numpy()
            labels = 'CONGRESOS'
        
            #media mensual
            mean_mes = df3[str(var)+'-CONGR'].T.to_numpy()
            std_mes = df4[str(var)+'-CONGR'].T.to_numpy()
        
            #plot
            
            fig1.plot(dias, mean_day, '.--', lw=0.5, label=labels)
            fig1.errorbar(mes, mean_mes, fmt='h', yerr=(std_mes), capsize=4, label=str(labels)+'-mes')
        else:
            #media diaria
            mean_day = df1[[str(var)+'-GRN', str(var)+'-CONGR']].T.to_numpy()
            std_day = df2[[str(var)+'-GRN', str(var)+'-CONGR']].T.to_numpy()
            labels = ['GR NORTE', 'CONGRESOS']
        
        
            #media mensual
            mean_mes = df3[[str(var)+'-GRN', str(var)+'-CONGR']].T.to_numpy()
            std_mes = df4[[str(var)+'-GRN', str(var)+'-CONGR']].T.to_numpy()
        
            #plot
            for i in range(len(mean_day)):
                fig1.plot(dias, mean_day[i], '.--', lw=0.5, label=labels[i])
                fig1.errorbar(mes, mean_mes[i], fmt='h', yerr=(std_mes[i]), capsize=4, label=str(labels[i])+'-mes')
            
    
    #Nombre de los ejes y save
    
    #VALORES LIMITE DIARIOS
    if(var=='PM10'):
        fig1.axhline(50, lw=0.2, c='r', label='Valor límite diario')
    
    fig1.set_xlabel('Fecha')
    
    if (var=='Temperature'):
        fig1.set_ylabel(str(var)+r'$[ºC]$')
    elif (var=='Humidity'):
        fig1.set_ylabel(str(var)+'[%]')
    elif (var=='Pressure'):
        fig1.set_ylabel(str(var)+'[mbar]')
    else:
        fig1.set_ylabel(str(var)+r'$[\mu g / m^3]$')
    
    fig1.legend()
    fig.tight_layout()
    plt.savefig(path)
```

```python
#FUNCION QUE ME OBTIENE LAS SERIES TEMPORALES DE UNA VARIABLE EN LAS DIFERENTES ESTACIONES
def serie2(df1, df2, var, delta):
    fig, fig1 = plt.subplots(1, 1, figsize=(12,6), dpi=200)
    dias = np.arange('2021-01-01', '2022-01-01', dtype='datetime64[D]')
    #CODIGO DE COLORES EN FUNCION DE LA ESTACION
    #CEAMA/LORCA/SDOMINGO
    colmes = ['purple', 'navy', 'forestgreen']
    #DISTINGO RED AIRLY DE JUNTA
    #red airly
    if (df1.columns[0].split('-')[1] in ['CEAMA', 'SDOMINGO', 'LORCA']):
        
        path = 'SERIES/AIRLY/serie_'+str(var)+'_'+str(delta)+'_AIRLY.eps'
        #Tengo en cuenta que no todas las variables estan en las tres estaciones
        if ((var=='CO') or (var=='SO2')):
            mean_day = df1[str(var)+'-SDOMINGO'].T.to_numpy()
            std_day = df2[str(var)+'-SDOMINGO'].T.to_numpy()
            labels = 'SDOMINGO'
            fig1.errorbar(df1.index, mean_day, fmt='h--', c=colmes[2], yerr=(std_day), capsize=2, label=labels)
    
        elif((var=='NO2') or (var=='O3')):
            mean = df1[[str(var)+'-CEAMA', str(var)+'-LORCA']].T.to_numpy()
            std = df2[[str(var)+'-CEAMA', str(var)+'-LORCA']].T.to_numpy()
            labels = ['CEAMA', 'LORCA']
            for i in range(len(mean)):
                fig1.errorbar(df1.index, mean[i], fmt='h--', c=colmes[i], yerr=(std[i]), capsize=2, label=labels[i])
    
        else:
            mean = df1[[str(var)+'-CEAMA', str(var)+'-LORCA', str(var)+'-SDOMINGO']].T.to_numpy()
            std = df2[[str(var)+'-CEAMA', str(var)+'-LORCA', str(var)+'-SDOMINGO']].T.to_numpy()
            labels = ['CEAMA', 'LORCA', 'SDOMINGO']
            for i in range(len(mean)):
                fig1.errorbar(df1.index, mean[i], fmt='h--', c=colmes[i], yerr=(std[i]), capsize=2, label=labels[i])
    #red junta
    else:
        path = 'SERIES/JUNTA/serie_'+str(var)+'_'+str(delta)+'_JUNTA.eps'
        
        if(var == 'O3'):
            mean = df1[str(var)+'-CONGR'].T.to_numpy()
            std = df2[str(var)+'-CONGR'].T.to_numpy()
            labels = 'CONGRESOS'
            fig1.errorbar(df1.index, mean, fmt='h--', yerr=(std), capsize=2, label=labels)
        else:
            mean = df1[[str(var)+'-GRN', str(var)+'-CONGR']].T.to_numpy()
            std = df2[[str(var)+'-GRN', str(var)+'-CONGR']].T.to_numpy()
            labels = ['GR NORTE', 'CONGRESOS']
            for i in range(len(mean)):
                fig1.errorbar(df1.index, mean[i], fmt='h--', yerr=(std[i]), capsize=2, label=labels[i])
    
    if (delta == 'semana'):
        fig1.set_xlabel('Dia de la semana')
    elif (delta == 'horas'):
        fig1.set_xlabel('Hora')
    else:
        fig1.set_xlabel('Fecha')
    
    if (var=='Temperature'):
        fig1.set_ylabel(str(var)+r'$[ºC]$')
    elif (var=='Humidity'):
        fig1.set_ylabel(str(var)+'[%]')
    elif (var=='Pressure'):
        fig1.set_ylabel(str(var)+'[mbar]')
    else:
        fig1.set_ylabel(str(var)+r'$[\mu g / m^3]$')
    
    fig1.legend()
    fig.tight_layout()
    plt.savefig(path)
```

```python
def series(df_list):
    list_serie = [medias(df_list[i]) for i in range(len(df_list))]
    
    #medias diarias
    medias_diarias = pd.concat([list_serie[i][0].T for i in range(len(df_list))]).T
    #std diarias
    std_diarias = pd.concat([list_serie[i][1].T for i in range(len(df_list))]).T
    #medias mensuales
    medias_mes = pd.concat([list_serie[i][2].T for i in range(len(df_list))]).T
    #std mensuales
    std_mes = pd.concat([list_serie[i][3].T for i in range(len(df_list))]).T
    #medias por dia de la semana
    media_semana = pd.concat([list_serie[i][4].T for i in range(len(df_list))]).T
    #std por dia de la semana
    std_semana = pd.concat([list_serie[i][5].T for i in range(len(df_list))]).T
    #media por horas
    media_horas = pd.concat([list_serie[i][6].T for i in range(len(df_list))]).T
    #std por horas
    std_horas = pd.concat([list_serie[i][7].T for i in range(len(df_list))]).T
    
    #plots
    if (df_list[0].columns[0].split('-')[1] == 'CEAMA'):
        variables = ['PM1', 'PM2.5', 'PM10', 'CO', 'O3', 'NO2', 'SO2', 'Temperature', 'Humidity', 'Pressure']
    else:
        variables = ['SO2', 'PM10', 'CO', 'O3', 'NO2']

    
    for i in range(len(variables)):
        with plt.style.context('ggplot'):
            #series anuales
            plot_año(medias_diarias, std_diarias, medias_mes, std_mes, variables[i])
            #series por dia de la semana
            serie2(media_semana, std_semana, variables[i], 'semana')
            #series por horas
            serie2(media_horas, std_horas, variables[i], 'horas')
    
    return print('Listo')
```

```python
#funcion que me separa los datos por estacion del año
def season(df):
    seasons = ['INV', 'PRI', 'VER', 'OTO']
    
    df_inv = df.loc[df.index <= '2021-04-01 00:00:00']
    df_pri = df.loc['2021-04-01 00:00:00':'2021-06-30 23:00:00']
    df_ver = df.loc['2021-07-01 00:00:00':'2021-09-30 23:00:00']
    df_oto = df.loc[df.index >= '2021-10-01 00:00:00']

    df_inv.columns = [df.columns[i].split('-')[0]+'-'+seasons[0] for i in range(len(df.columns))]
    df_pri.columns = [df.columns[i].split('-')[0]+'-'+seasons[1] for i in range(len(df.columns))]
    df_ver.columns = [df.columns[i].split('-')[0]+'-'+seasons[2] for i in range(len(df.columns))]
    df_oto.columns = [df.columns[i].split('-')[0]+'-'+seasons[3] for i in range(len(df.columns))]
    
    return df_inv, df_pri, df_ver, df_oto
```

```python
from scipy.stats import pearsonr
import pandas as pd

def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    df.fillna(0,inplace=True)
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
    return pvalues
```

```python
def media_octoh(df,var,est, h):
    dias = pd.date_range(start='2021-01-01 00:00:00', end='2022-01-01 00:00:00', freq='D').strftime('%Y-%m-%d')
    
    serie = df[str(var)+'-'+str(est)].rolling(window=h).mean()
    
    max_dia = [serie[dias[i]].max() for i in range(len(dias)-1)]
    
    df_max = pd.DataFrame(max_dia, index=dias[0:-1], columns=[str(var)+'-'+str(est)])
    return df_max
```

```python
def supera(var):
    red_airly = [df_ceama, df_sdomingo, df_lorca]
    red_junta = [df_gr_n, df_congr]
    
    
    est_airly = ['CEAMA', 'SDOMINGO', 'LORCA']
    est_junta = ['GRN', 'CONGR']
    if var=='PM10':
        for i in range(len(est_airly)):
            dias = pd.read_excel('DATOS/AIRLY/medias_diarias_'+str(est_airly[i])+'.xlsx')
            sup = dias[str(var)+'-'+est_airly[i]]>50
            nsup = len(sup[sup==True])
            print('Superaciones de '+str(var)+' en '+str(est_airly[i])+':'+str(nsup))

        for i in range(len(est_junta)):
            dias = pd.read_excel('DATOS/JUNTA/medias_diarias_'+str(est_junta[i])+'.xlsx')
            sup = dias[str(var)+'-'+est_junta[i]]>50
            nsup = len(sup[sup==True])
            print('Superaciones de '+str(var)+' en '+str(est_junta[i])+':'+str(nsup))
    elif var=='SO2':
        dias = pd.read_excel('DATOS/AIRLY/medias_diarias_'+str(est_airly[1])+'.xlsx')
        sup = dias[str(var)+'-'+est_airly[1]]>125
        nsup = len(sup[sup==True])
        print('Superaciones de '+str(var)+' en '+str(est_airly[1])+':'+str(nsup))

        for i in range(len(est_junta)):
            dias = pd.read_excel('DATOS/JUNTA/medias_diarias_'+str(est_junta[i])+'.xlsx')
            sup = dias[str(var)+'-'+est_junta[i]]>125
            nsup = len(sup[sup==True])
            print('Superaciones de '+str(var)+' en '+str(est_junta[i])+':'+str(nsup))
    elif var=='O3':
        for i in range(0, len(est_airly), 2):
            dias = media_octoh(red_airly[i], var, est_airly[i], 8)
            sup = dias[str(var)+'-'+str(est_airly[i])]>120
            nsup = len(sup[sup==True])
            print('Superaciones de '+str(var)+' en '+str(est_airly[i])+':'+str(nsup))

        dias = media_octoh(red_junta[1], var, est_junta[1], 48)
        sup = dias[str(var)+'-'+str(est_junta[1])]>120
        nsup = len(sup[sup==True])
        print('Superaciones de '+str(var)+' en '+str(est_junta[1])+':'+str(nsup))
    elif var=='CO':
        for i in range(0, len(est_junta)):
            dias = media_octoh(red_junta[i], var, est_junta[i], 48)
            sup = dias[str(var)+'-'+str(est_junta[i])]>10e+03
            nsup = len(sup[sup==True])
            print('Superaciones de '+str(var)+' en '+str(est_junta[i])+':'+str(nsup))

        dias = media_octoh(red_airly[1], var, est_airly[1], 48)
        sup = dias[str(var)+'-'+str(est_airly[1])]>10e+03
        nsup = len(sup[sup==True])
        print('Superaciones de '+str(var)+' en '+str(est_airly[1])+':'+str(nsup))
variables = ['PM10', 'SO2', 'O3', 'CO']
for i in range(len(variables)):
    supera(variables[i])
```

```python
def load_hourly():
    df = pd.read_excel('DATOS/AIRLY/estadistica_horas_año_CEAMA.xlsx', index_col=[0,1])
    medias_horas = df.xs('mean', level=1).T
    std_horas = df.xs('std', level=1).T
    
    means_pm = medias_horas[['PM1-CEAMA', 'PM2.5-CEAMA', 'PM10-CEAMA']]
    std_pm = std_horas[['PM1-CEAMA', 'PM2.5-CEAMA', 'PM10-CEAMA']]
    with plt.style.context('ggplot'):
        fig, fig1 = plt.subplots(figsize=(12,6))
        plt.rc('legend',fontsize=20)
        means_pm.plot(yerr=std_pm, ax=fig1, fmt='h--', capsize=4)
        fig1.set_xlabel('Hour', fontsize=24)
        fig1.set_ylabel('PM concentration'+r'$[\mu g / m^3]$', fontsize=24)
        fig1.set_ylim(0,70)
        plt.xticks(fontsize=20)
        plt.yticks(fontsize=20)
        fig.tight_layout()
    
    means_gas = medias_horas[['NO2-CEAMA', 'O3-CEAMA']]
    std_gas = std_horas[['NO2-CEAMA', 'O3-CEAMA']]
    with plt.style.context('ggplot'):
        fig, fig2 = plt.subplots(figsize=(12,6))
        plt.rc('legend',fontsize=20)
        plt.xticks(fontsize=20)
        plt.yticks(fontsize=20)
        ax2 = fig2.twinx()
        means_gas['NO2-CEAMA'].plot(yerr=std_gas['NO2-CEAMA'], ax=fig2, fmt='h--', color='crimson', label='NO2', capsize=4)
        means_gas['O3-CEAMA'].plot(yerr=std_gas['O3-CEAMA'], ax=ax2, fmt='h--', color='navy', label='O3', capsize=4)
        fig2.set_xlabel('Hour', fontsize=24)
        fig2.set_ylabel('NO2 concentration'+r'$[\mu g / m^3]$', fontsize=24)
        ax2.set_ylabel('O3 concentration'+r'$[\mu g / m^3]$', fontsize=24)
        fig2.legend(loc=2)
        ax2.legend(loc=1)
        plt.yticks(fontsize=20)
        fig.tight_layout()
load_hourly()
```

```python
def load_season(df):
    df = df[['PM1-CEAMA', 'PM2.5-CEAMA', 'PM10-CEAMA', 'NO2-CEAMA', 'O3-CEAMA']]
    est = ['1WIN', '2SPR', '3SUM', '4AUT']
    col = [df.columns[i].split('-')[0] for i in range(len(df.columns))]
    list_df = season(df)
    
    for i in range(len(list_df)):
        list_df[i].columns = pd.MultiIndex.from_product([col, [est[i]]])
        
    df_seasons = pd.concat(list_df)
    
    df_seasons = df_seasons[['PM1', 'PM2.5', 'PM10', 'NO2', 'O3']]
    
    fig, fig1 = plt.subplots(1, 1, figsize=(12,6))
    plt.rc('legend',fontsize=20)
    plt.yticks(fontsize=20)
    
    pos = [1, 2, 3, 4, 6, 7, 8, 9, 11, 12, 13, 14, 16, 17, 18, 19, 21, 22, 23, 24]
    df_seasons.boxplot(rot=0, positions=pos, ax=fig1, showmeans=True, sym='', grid=False)
    fig1.axvline(5, color='k', lw=0.5)
    fig1.axvline(10, color='k', lw=0.5)
    fig1.axvline(15, color='k', lw=0.5)
    fig1.axvline(20, color='k', lw=0.5)
    
    fig1.set_xticks(pos)
    fig1.set_xticklabels(['WIN', 'SPR', 'SUM', 'AUT',
                          'WIN', 'SPR', 'SUM', 'AUT', 
                          'WIN', 'SPR', 'SUM', 'AUT', 
                          'WIN', 'SPR', 'SUM', 'AUT',
                          'WIN', 'SPR', 'SUM', 'AUT'], fontsize=15)
    ax = fig1.twiny()
    ax.set_xticks([2.5, 8, 13.75, 19.5, 25, 27.5])
    ax.set_xticklabels(['PM1', 'PM2.5', 'PM10', 'NO2', 'O3', ' '], fontsize=24)
    
    fig1.set_ylabel('Pollutants concentration'+r'$[\mu g / m^3]$', fontsize=24)
    plt.grid(False)
    fig.tight_layout()
    
    return df_seasons
with plt.style.context('ggplot'):
    load_season(df_ceama)
```
